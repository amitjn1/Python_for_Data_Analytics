```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python

# Step 1: NumPy - Creating and manipulating arrays
data = np.random.randn(1000)
data
```

```python
# Step 2: Pandas - Creating a DataFrame
df = pd.DataFrame(data, columns=['Value'])
df
```

```python
# Step 3: Matplotlib - Basic visualization
plt.hist(df['Value'], bins=30)
plt.title('Histogram of Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Step 4: Seaborn - Enhanced visualization
sns.histplot(df['Value'], bins=30, kde=True)
plt.title('Seaborn Histogram of Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
